```python
import sqlite3
import pandas as pd
from utils.bot import PetDB
```

```python
db = PetDB(r"data/dyrevaernet.db")
```

```python
# join pet and type tables
pd.read_sql("""
    SELECT pet.name, race.name AS race, type.name AS type, status.value as status
    FROM pet
    INNER JOIN race
    ON pet.race_id = race.id
    INNER JOIN type
    ON pet.type_id = type.id
    INNER JOIN status
    ON pet.status_id = status.id
    WHERE type.name = 'cat'""", 
    con=db.conn)
```

```python
# READ ALL FROM PET TABLE
pd.read_sql("SELECT * FROM pet", con=db.conn, index_col='id')
```
